```python
#Baixando o arquivo, lembre de alterar..
import pandas as pd

path = "Desktop/data_science_alura/"
dataset = pd.read_csv("{}dados_eyetracking.csv".format(path), sep = ';', index_col = 0)
dataset=dataset.reset_index()

#criando uma coluna que afirmar se um dado é uma duplicata
dataset2 = dataset.iloc[:,[0,3]]
dataset['duplicados'] = (dataset2.duplicated())

#Identifica a duplicata e faz a média das duas coletas
for i in range(1, len(dataset)):
     if dataset.iloc[i]['duplicados'] == True:
           dataset.iloc[i-1,5:30] = dataset.iloc[i-1:i+1,5:30].mean()
        
#Exclui as linhas duplicadas a coluna que identifica se a a linha é duplicada
dataset = dataset[dataset.duplicados != True]
dataset = dataset.drop('duplicados', 1)   

#exporta para excel
dataset.to_csv("{}dados_eyetracking_editados.csv".format(path), index = True, header=True)
dataset
```
